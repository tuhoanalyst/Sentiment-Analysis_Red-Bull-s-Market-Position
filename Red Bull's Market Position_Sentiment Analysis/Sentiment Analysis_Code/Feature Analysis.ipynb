# The following code is used to generate comparison table of features for Red Bull and other three brands. Stacked bar graph will be created by excel. 


```python
import pandas as pd

rb_df = pd.read_csv('feature red bull.csv')
mons_df = pd.read_csv('feature monster.csv')
c4_df = pd.read_csv('feature c4.csv')
alani_df = pd.read_csv('feature alani.csv')
```

```python
mons_df['sentiment'].value_counts()
```

```python
rb_df['date'] = pd.to_datetime(rb_df['date'])
mons_df['date'] = pd.to_datetime(mons_df['date'])
c4_df['date'] = pd.to_datetime(c4_df['date'])
alani_df['date'] = pd.to_datetime(alani_df['date'])
```

```python
average_scores = {
    'Brand': ['Red Bull', 'Monster Energy', 'C4', 'Alani'],
    'Taste': [
        rb_df[rb_df['date'].dt.year == 2024]['sentiment_taste'].mean(),
        mons_df[mons_df['date'].dt.year == 2024]['sentiment_taste'].mean(),
        c4_df[c4_df['date'].dt.year == 2024]['sentiment_taste'].mean(),
        alani_df[alani_df['date'].dt.year == 2024]['sentiment_taste'].mean()
    ],
    'Flavor': [
        rb_df[rb_df['date'].dt.year == 2024]['sentiment_flavor'].mean(),
        mons_df[mons_df['date'].dt.year == 2024]['sentiment_flavor'].mean(),
        c4_df[c4_df['date'].dt.year == 2024]['sentiment_flavor'].mean(),
        alani_df[alani_df['date'].dt.year == 2024]['sentiment_flavor'].mean()
    ],
    'Price': [
        rb_df[rb_df['date'].dt.year == 2024]['sentiment_price'].mean(),
        mons_df[mons_df['date'].dt.year == 2024]['sentiment_price'].mean(),
        c4_df[c4_df['date'].dt.year == 2024]['sentiment_price'].mean(),
        alani_df[alani_df['date'].dt.year == 2024]['sentiment_price'].mean()
    ],
    'Packaging': [
        rb_df[rb_df['date'].dt.year == 2024]['sentiment_packaging'].mean(),
        mons_df[mons_df['date'].dt.year == 2024]['sentiment_packaging'].mean(),
        c4_df[c4_df['date'].dt.year == 2024]['sentiment_packaging'].mean(),
        alani_df[alani_df['date'].dt.year == 2024]['sentiment_packaging'].mean()
    ],
    'Value': [
        rb_df[rb_df['date'].dt.year == 2024]['sentiment_value'].mean(),
        mons_df[mons_df['date'].dt.year == 2024]['sentiment_value'].mean(),
        c4_df[c4_df['date'].dt.year == 2024]['sentiment_value'].mean(),
        alani_df[alani_df['date'].dt.year == 2024]['sentiment_value'].mean()
    ],
    'Package': [
        rb_df[rb_df['date'].dt.year == 2024]['sentiment_package'].mean(),
        mons_df[mons_df['date'].dt.year == 2024]['sentiment_package'].mean(),
        c4_df[c4_df['date'].dt.year == 2024]['sentiment_package'].mean(),
        alani_df[alani_df['date'].dt.year == 2024]['sentiment_package'].mean()
    ]
}

scores_df_2024 = pd.DataFrame(average_scores)

scores_df_2024

```
